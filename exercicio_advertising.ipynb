# Usando apenas TV e Radio 

```python
import pandas as pd
publi = pd.read_csv('Advertising.csv', index_col=0)
publi.head()
```

```python
x_tv_rad = publi[['TV', 'Radio']]
x_tv_news = publi[['TV', 'Newspaper']]
x_rad_news = publi[['Radio', 'Newspaper']]

y = publi['Sales']
```

```python
from sklearn.model_selection import train_test_split
x_train_tvrad, x_test_tvrad, y_train_tvrad, y_test_tvrad = train_test_split(x_tv_rad, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression 
reglin = LinearRegression()
reglin.fit(x_train_tvrad, y_train_tvrad)
```

```python
y_prev_tvrad = reglin.predict(x_test_tvrad)
print(y_prev_tvrad)
```

```python
# RMSE
from sklearn import metrics
import numpy as np
print(f'Usando apenas TV e Radio --> Apenas um erro de {np.sqrt(metrics.mean_squared_error(y_test_tvrad, y_prev_tvrad))}')
```
